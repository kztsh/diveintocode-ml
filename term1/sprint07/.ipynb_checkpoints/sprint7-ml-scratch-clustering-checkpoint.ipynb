## 1. K-meansクラスの作成
非階層クラスタリングの手法であるK-meansのクラスを作成する。名前はScratchKMeansとする。  
教師なし学習であるクラスタリングになるため、ここまでの回帰や分類とは異なるパイプラインを作成することになる。  
メソッドやパラメータなどはscikit-learnを参考にする。  
[sklearn.cluster.KMeans — scikit-learn 0.20.0 documentation](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html "sklearn.cluster.KMeans — scikit-learn 0.20.0 documentation")

検証段階ではJupyter Notebook上で書いていき、最終的にはpyファイルにモジュールとしてまとめる。  

**クラスタリングのための人工データセット**  
クラスタリングを実験するための人工データセット（シンプルデータセット3）を作成するコードをスクラッチの過程で利用する。  
このmake_blobs関数は正解ラベルも出力するが、今回は使用しない。使用しないことを明示するために、 \_（アンダースコア） で受け取っている。

**目的関数**  
K-meansは以下のクラスタ内誤差平方和（SSE, Sum of Squared Errors）を最小化する$r_{nk}$と$\mu_k$を求めることで学習を行う。  
この数式自体は学習時にはコーディングする必要がないが、クラスタ数$K$を決めるためのエルボー法で使う。  
$$SSE = \sum_{n=1}^N \sum_{k=1}^K r_{nk} \|X_n - \mu_k\|^2$$

$n$  : データ点のインデックス  
$k$ : クラスタのインデックス  
$X_n$ : $n$番目のデータ点  
$\mu_k$ : $k$番目の中心点  
$r_{nk}$ : データ点$X_n$がクラスタ$k$に所属していたら1、そうでなければ0  

**学習**  
K-meansの学習の基本は以下の2つのフェーズを繰り返す。

- 中心点$\mu_k$を固定した上で$J$を最小化する$r_{nk}$を選ぶ。
- $r_{nk}$を固定した上で$J$を最小化する$\mu_k$を選ぶ。

具体的には以下を行う。  
まずは  
- 中心点$\mu_k$の初期値を決める。  

その後以下を繰り返す。

- データ点$X_n$を最も近い$\mu_k$に割り当てる。
- 中心点$\mu_k$を$k$番目のクラスタに割り当てられる全てのデータ点$X_n$の平均値（重心）に移動する。

次のどれかの条件を満たしたらこれを止める。

- 繰り返しが指定した回数に到達する。
- 中心点$\mu_k$と重心の差が指定した許容値以下になる。
- データ点$X_n$のクラスタへの割り当てが変化しなくなる。

さらに、

- ここまでの一連の流れを指定した回数繰り返す。
- SSEが最小になったものを最終的な学習結果とする。

これで学習が完了する。

考慮すべき点として、空のクラスタが生じる可能性がある。これを防ぐためには以下の動作を実装する。
$k$番目のクラスタに割り当てられるデータ点$X_n$が存在しない場合、中心点$\mu_k$を最も離れているデータ点の場所に移動する。

繰り返し回数や許容値などのハイパーパラメータの値はscikit-learnのデフォルト値を参考にする。

**ユークリッド距離**  
K-menasにおける近いとは、点と点のユークリッド距離が小さくなること。  
ユークリッド距離とはピタゴラスの定理（三平方の定理）で求められるが、ベクトル$p,q$に対しては以下の数式で表現できる。
$$\|q-p\| = \sqrt{(q-p)\cdot(q-p)}$$
NumPyにはこの関数がnp.linalg.normとして用意されているため使用する。

**参考**  
[numpy.linalg.norm — NumPy v1.15 Manual](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.linalg.norm.html "numpy.linalg.norm — NumPy v1.15 Manual")

**初期値**  
K-meansの学習結果は中心点$\mu$の初期値に影響を受ける。  
最もシンプルな初期値の決定方法は、データ点$X_n$の中からランダムに$k$個選ぶこと。今回はこれを実装する。  
K-meansのより良い初期値の選択方法については研究が行われており、いろいろな方法が提案されている。  

**推定**  
新たなデータに対してどのクラスタに所属するかの推定の際は、学習時の1つ目のフェーズと同じことを行う。

- データ点$X_n$に対し、学習で決まった中心点$\mu_k$の元、$J$を最小化する$r_{nk}$を選ぶ。

具体的には次のこと。
- データ点$X_n$を最も近い$\mu_k$に割り当てる。

```python
# シンプルデータセット3
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=100, n_features=2, centers=4, cluster_std=0.5, shuffle=True, random_state=0)
```

```python
import sys
sys.path.append("../ml-scratch/models/")
import pipe_clf
from importlib import reload
reload(pipe_clf)

import numpy as np
import matplotlib.pyplot as plt

class ScratchKMeans:
    """
    KMeansのスクラッチ実装(Xは2次元特徴量を想定)
    
    距離計算(dist変数)では、
    1. Xをaxis=0,1に、cluster_centersをaxis=1,2に、それぞれ3次元空間に展開
    2. そのブロードキャスト引算値の２乗を計算
    3. axis=1方向、つまり(x1_data - x1_clusters)**2と(x2_data - x2_clusters)**2で和をとる
    の3ステップにより、元のaxis=0（データ数）方向と元のaxis=2（cluster_centers数）方向の行列を返している。
    行列の各要素は各データ点と各クラスター中心とのユークリッド距離の2乗値が入っている。
    
    _cluster_func内部の「イテレーション」によるクラスタ内誤差平方和（SSE: Sum of Squared Errors）の最小化をn_iter回「繰り返し」、
    SSE最小値のうち最も小さい値の場合に以下のAttributesを保存する。

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        学習用データの特徴量
    n_clusters : int
        クラスターの数
    n_iter : int (default : 10)
        異なる乱数にてSSE最小化を繰り返し、その中で最小のSSEを求める「繰り返し」回数
    max_iter : int (default : 100)
        SSE最小化のための「イテレーション」の最大回数
    threshold : float (default : 1e-5)
        中心点と重心の差の許容値
    random_state : int (default : 0)
        初期ラベルのシャッフル、cluster_centersの初期値設定をするための擬似乱数シード値
    
    Attributes
    ----------
    self.labels_ : ndarray, shape(n_samples, )
      Xの各サンプル毎の予測ラベル値
    self.cluster_centers_ : ndarray, shape (n_centers, n_features)
      各クラスター中心座標
    self.sse_history_ : list
      SSEが最小値をとった際のイテレーションにおけるSSEの履歴値
    self.sse_ : float
      self.sse_historyにおける最後の値（SSE最小値）
    """
    def __init__(self, n_clusters, n_iter=10, max_iter=100, threshold = 1e-5, random_state=0):
        self.n_clusters = n_clusters
        self.n_iter = n_iter
        self.max_iter = max_iter
        self.threshold = threshold
        self.rgen = np.random.RandomState(random_state)
        self.sse_history_ = [np.inf]
    
    def fit(self, X):
        # n_iter回数分繰り返す
        for _ in range(self.n_iter):
            self._cluster_func(X)
        # SSEの履歴のうち、最終の値を格納する
        self.sse_ = self.sse_history_[-1]
        
    def _cluster_func(self, X):
        # 必要な初期値を生成
        pre_labels = np.zeros(X.shape[0])
        iteration = 0
        
        # サンプル数分のインデックス配列を作り、クラスタ数で割った余りを初期ラベルとする(3クラスなら[0,1,2,0,1,2,0, ...])
        n_samples = np.arange(X.shape[0])
        labels = np.array([x%self.n_clusters for x in n_samples], dtype=np.int)
        # 初期ラベルを乱数ジェネレータでシャッフル
        self.rgen.shuffle(labels)
        
        # 更新前後のcluster_centersをXのサンプルからランダムに選択
        pre_cluster_centers = X[self.rgen.randint(0, X.shape[0], self.n_clusters), :]
        cluster_centers = X[self.rgen.randint(0, X.shape[0], self.n_clusters), :]
        
        # 上記の両者の差の2乗和の初期値を生成
        cluster_centers_diff = ((cluster_centers - pre_cluster_centers)**2).sum(axis=1)
        
        # cluster_centersと各データ点との距離を初期値として計算
        dist = ((X[:, :, np.newaxis] - cluster_centers.T[np.newaxis, :, :])**2).sum(axis=1)
        # SSEの初期値を計算
        sse = 0
        sse_history = []
        for i in range(dist.shape[1]):
            sse += dist[labels==i, i].sum()
        sse_history.append(sse)
        
        
        # 以下から学習のイテレーションに入る
        """
        以下3パターンのどれかを満たしたら更新ストップ
        1. ラベル更新が全サンプル点で止まる
        2. 中心点と重心の間の距離の2乗（更新前後のcluster_centersの差）が指定した許容値以下になる
        3. イテレーション回数がmax_iterに至る
        """
        while (not ((labels == pre_labels).all()) and 
               not ((cluster_centers_diff < self.threshold).all()) and 
               iteration < self.max_iter):
            
            # 最初に現時点でのラベル値とクラスター中心値を保持
            pre_labels = labels.copy()
            pre_cluster_centers = cluster_centers.copy()
            
            # 各ラベルに該当するデータ点毎の平均値（重心）でcluster_centersを更新
            unique_labels = np.unique(labels)
            for i in range(self.n_clusters):
                # 空のクラスタの場合、クラスタ中心を最も離れているデータ点の場所に移動する
                if i not in unique_labels:
                    j = np.argmax(dist[:, i]) # distのi列目（i番目のcluster_center列）から最大の行番号をとる
                    cluster_centers[i, :] = X[j, :] # その行番号に該当するXの行データで更新する
                # 空のクラスタでない場合は平均値（重心）で更新
                else:
                    cluster_centers[i, :] = X[labels == i, :].mean(axis=0)
            # 更新前後のcluster_centersの差を更新
            cluster_centers_diff = ((cluster_centers - pre_cluster_centers)**2).sum(axis=1)
            
            # 更新されたcluster_centersと各データ点との距離を計算
            dist = ((X[:, :, np.newaxis] - cluster_centers.T[np.newaxis, :, :])**2).sum(axis=1)
            # 最短距離のラベルインデックスを新たなラベルとする
            labels = np.argmin(dist, axis=1)
            
            # SSEを計算
            sse = 0
            for i in range(dist.shape[1]):
                sse += dist[labels==i, i].sum()
            sse_history.append(sse)
            
            # イテレーション回数を更新
            iteration += 1
        
        # インスタンス変数と比較し、最終のSSEが小さければインスタンス変数を更新
        if self.sse_history_[-1] > sse_history[-1]:
            self.sse_history_ = sse_history
            self.labels_ = labels
            self.cluster_centers_ = cluster_centers        
    
    def predict(self, X):
        # self.cluster_centers_と各データ点との距離を計算
        dist = ((X[:, :, np.newaxis] - self.cluster_centers_.T[np.newaxis, :, :])**2).sum(axis=1)
        
        # 最短距離のラベルインデックスを予測ラベルとする
        labels = np.argmin(dist, axis=1)
        return labels
    
```

```python
# スクラッチ版のインスタンスを生成して学習
skm = ScratchKMeans(n_clusters=4)
skm.fit(X)

# 決定領域の描画
plt.rcParams["font.size"] = 16
fig, ax = plt.subplots(1,2,figsize=(20,6))
plt.subplot(1,2,1)
pipe_clf.decision_region(X, skm.predict(X), skm, model_name="ScratchKMeans")

# SSEの描画
plt.subplot(1,2,2)
plt.plot(np.arange(len(skm.sse_history_)), skm.sse_history_, marker="o")
plt.xticks(np.arange(len(skm.sse_history_)))
plt.xlabel("Iterations")
plt.ylabel("SSE")
plt.show()
```

## 2. エルボー法の実装
K-meansにおいて中心点の個数$k$は人為的に決める必要があるが、その判断方法としてエルボー法がある。  
SSEを縦軸、クラスタ数$k$を横軸にした折れ線グラフを表示する。  
このためにK-meansのインスタンスからはSSEを取得できるようにする。  

```python
sse_list = []
clusters_range = range(1,11)
for i in clusters_range:
    skm = ScratchKMeans(n_clusters=i)
    skm.fit(X)
    sse_list.append(skm.sse_)

plt.plot(clusters_range, sse_list, marker="o")
plt.xticks(np.arange(1, len(clusters_range)+1))
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.title("Elbow method")
plt.show()
```

## 3. クラスタ数kの選択
**卸売業者のデータ分析**  
クラスタリングを利用法を学ぶために、Wholesale customers Data Setを使用する。

[UCI Machine Learning Repository: Wholesale customers Data Set](https://archive.ics.uci.edu/ml/datasets/Wholesale+customers "UCI Machine Learning Repository: Wholesale customers Data Set")

Data Folderの中からWholesale customers data.csvをダウンロードする。  
これはポルトガルのリスボンにある卸売業者（wholesale）が顧客から集めたデータである。  
中身は各顧客が年間でミルクや食料品を何ユーロ購入しているかというデータになっている。  
顧客は小売店やレストランである。  
RegionとChannelは取り除き、金額のデータでK-meansを利用する。  

**次元削減**  
金額に関する特徴量は6個あるが、可視化するために次元削減を行う。  

最適だと思われるクラスタ数$k$を選択する。  
エルボー法やシルエット図を活用する他、人間が後々どう利用したいのかも選択に関わってくる。  
選択した上で、なぜそれを選んだか説明する。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools
from sklearn.decomposition import PCA

# 小数点表示は2桁まで
pd.options.display.precision = 2

# データ読み込み
data = pd.read_csv("Wholesale customers data.csv")

# 顧客ID番号列を追加
customer_id = pd.DataFrame(np.arange(data.shape[0]), columns=["Customer ID"])
data = pd.concat((customer_id, data), axis=1)

# 欠損値の確認
print("欠損値の有無：", data.isnull().sum().sum())

# データの外観
data.head()
```

```python
# PCAの分散説明率を表示
pca = PCA(n_components = None)
pca = pca.fit(data.iloc[:, 3:])
exp_var = pca.explained_variance_ratio_
print("Explained variance: ", exp_var)
cum_exp_var = np.cumsum(exp_var)
print("Cumulative explained variance: ", cum_exp_var)
plt.figure(figsize=(10,6))
plt.bar(range(1,7), exp_var, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1,7), cum_exp_var, where='mid', label='cumulative expalined variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.hlines(0.7, 0, 6,  "blue", linestyles='dashed')
plt.legend(loc='best')
plt.grid()
plt.show()

# PCAによる次元削減
pca = PCA(n_components = 2)
data_pca = pca.fit_transform(data.iloc[:, 3:])
plt.scatter(data_pca[:, 0], data_pca[:, 1], s=80, alpha=0.5, marker='o')
plt.xlabel("pca_components_0")
plt.ylabel("pca_components_1")
plt.show()
```

以下で、エルボー法、シルエット図、実際のスクラッチクラスタリングで最適なkを検討してみる。

```python
# エルボー法で最適なkを選択
# データは2つ上のセルのdata_pca(2_components)を使用
sse_list = []
clusters_range = range(1,11)
for i in clusters_range:
    skm = ScratchKMeans(n_clusters=i)
    skm.fit(data_pca)
    sse_list.append(skm.sse_)

plt.plot(clusters_range, sse_list, marker="o")
plt.xticks(np.arange(1, len(clusters_range)+1))
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.title("Elbow method")
plt.show()
```

```python
# シルエット図を表示
from matplotlib import cm
from sklearn.metrics import silhouette_samples

# ラベル1つの場合はシルエット係数が算出できないので除外
clusters_range = range(2,11)
for i in clusters_range:
    skm = ScratchKMeans(n_clusters=i)
    skm.fit(data_pca)
    y_km = skm.predict(data_pca)
    cluster_labels = np.unique(y_km)
    n_clusters = cluster_labels.shape[0]
    
    # シルエット係数を計算
    silhouette_vals = silhouette_samples(data_pca, y_km, metric="euclidean")
    y_ax_lower, y_ax_upper = 0, 0
    yticks = []
    # シルエット係数をラベル毎に描画
    for j, c in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[y_km == c]
        c_silhouette_vals.sort()
        y_ax_upper += len(c_silhouette_vals)
        color = cm.jet(j / n_clusters)
        # シルエット係数を水平棒として表示
        plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, 
                 height=1.0, edgecolor='none', color=color)
        yticks.append((y_ax_lower + y_ax_upper) / 2)
        y_ax_lower += len(c_silhouette_vals)
    
    silhouette_avg = np.mean(silhouette_vals)
    plt.axvline(silhouette_avg, color="red", linestyle="--")
    plt.yticks(yticks, cluster_labels + 1)
    plt.ylabel('Cluster')
    plt.xlabel('Silhouette coefficient')
    plt.title("{} labels".format(i))
    plt.show()
```

```python
# スクラッチ版のインスタンスを生成して学習
clusters_range = range(1,11)
for i in clusters_range:
    skm = ScratchKMeans(n_clusters=i)
    skm.fit(data_pca)
    
    # 決定領域の描画
    fig, ax = plt.subplots(1,2,figsize=(20,6))
    plt.subplot(1,2,1)
    pipe_clf.decision_region(data_pca, skm.predict(data_pca), skm, 
                             model_name="ScratchKMeans (k={})".format(i), 
                            feature_names=["pca_components_0","pca_components_1"])
    
    # SSEの描画
    plt.subplot(1,2,2)
    plt.plot(np.arange(len(skm.sse_history_)), skm.sse_history_, marker="o")
    plt.xticks(np.arange(len(skm.sse_history_)))
    plt.xlabel("Iterations")
    plt.ylabel("SSE")
    plt.title("SSE curve (k={})".format(i))
    plt.show()
```

## 5. 卸売業者にとって有益な情報
卸売業者にとって有益な情報を探し出し文章にまとめる。  
前処理を行う前のデータを使い、各クラスタの要約統計量などを見る。  
また、PCAで次元削減したデータに対してK-meansを使用したが、6個の特徴量をそのまま使ったり、一部の特徴量のみを使うことも試す。  
既知のグループとの比較結果も参考にする。  
条件ごとにデータを見る際にはPandasのgroupbyが便利となる。  

[pandas.DataFrame.groupby — pandas 0.23.4 documentation](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html "pandas.DataFrame.groupby — pandas 0.23.4 documentation")

```python
# 特徴量6つから2つを取ってくるインデックスペアで、生データ(deldata)を可視化
print("------Channel特徴量列のラベル別で可視化------")
combs = list(itertools.combinations(np.arange(6), 2))
# 生データの連続値特徴量を別の変数に格納
deldata_numerical = deldata.iloc[:, 3:]

# deldataで可視化
plt.subplots(5,3,figsize=(20,20))
for i, comb in enumerate(combs):
    plt.subplot(5,3,i+1)
    selected_data = deldata_numerical.iloc[:, [comb[0], comb[1]]]
    columns = deldata_numerical.columns[[comb[0], comb[1]]]
    # "Channel"列の値別にプロットの色を分ける
    plt.scatter(selected_data.iloc[:, 0][deldata["Channel"]==1], 
                selected_data.iloc[:, 1][deldata["Channel"]==1], 
                color="steelblue", s=80, alpha=0.5, marker='o', label="Horeca")
    plt.scatter(selected_data.iloc[:, 0][deldata["Channel"]==2], 
                selected_data.iloc[:, 1][deldata["Channel"]==2], 
                color="orangered", s=80, alpha=0.5, marker='o', label="Retail")
    plt.xlabel(columns[0])
    plt.ylabel(columns[1])
    plt.legend(loc="best")

plt.tight_layout()
plt.show()
```

凡例の意味は以下の通り。  
- Horeca: ホテル、レストラン、カフェの店舗の発注金額
- Retail: 小売業者の発注金額

店舗規模の違いによる絶対的な発注金額の大小以外に、下記の特徴があることがわかる。
- HorecaではFrozenに対してMilkの発注金額が低く、Retailは逆の傾向がある。
- RetailではFrozenに対してGroceryの発注金額が圧倒的に多い
- HorecaではDelicassenに対してDetergents_Paperの発注が少ない

次に、pca_componentsの2つの特徴量をfittingに用い、クラスター数毎(k=2から4)の予測ラベルに応じた発注品目別の発注金額をプロットする。

```python
# クラスター数毎(k=2から4)に予測ラベルを格納した配列を生成
labels_per_clusters = []
clusters_range = range(2,5)
for i in clusters_range:
    skm = ScratchKMeans(n_clusters=i)
    skm.fit(data_pca)
    labels_per_clusters.extend(skm.labels_)

labels_per_clusters = np.array(labels_per_clusters).reshape(3,421)
```

```python
# kの値毎にプロット
left = np.arange(6)+1
labels = deldata.columns[3:]

# k=2の場合
plt.subplots(1,2,figsize=(20,10))
for i in range(2):
    plt.subplot(1,2,i+1)
    selected_data = deldata[labels_per_clusters[0, :]==i].sum(axis=0)
    plt.bar(left, selected_data[3:], color="steelblue", alpha=0.5)
    plt.xticks(left, labels, rotation=90)
    plt.xlabel("Features")
    plt.ylabel("Annual spending(€)")
    plt.title("cluster label {} (k=2)".format(i))
plt.tight_layout()
plt.show()
    
# k=3の場合
plt.subplots(1,3,figsize=(20,10))
for i in range(3):
    plt.subplot(1,3,i+1)
    selected_data = deldata[labels_per_clusters[1, :]==i].sum(axis=0)
    plt.bar(left, selected_data[3:], color="orangered", alpha=0.5)
    plt.xticks(left, labels, rotation=90)
    plt.xlabel("Features")
    plt.ylabel("Annual spending(€)")
    plt.title("cluster label {} (k=3)".format(i))
plt.tight_layout()
plt.show()

# k=4の場合
plt.subplots(2,2,figsize=(20,20))
for i in range(4):
    plt.subplot(2,2,i+1)
    selected_data = deldata[labels_per_clusters[2, :]==i].sum(axis=0)
    plt.bar(left, selected_data[3:], color="lightgreen", alpha=0.5)
    plt.xticks(left, labels, rotation=90)
    plt.xlabel("Features")
    plt.ylabel("Annual spending(€)")
    plt.title("cluster label {} (k=4)".format(i))
plt.tight_layout()
plt.show()
```

分類クラスタ別に特徴量毎の発注金額が異なり、顧客の層に応じて発注量を調整する目安のデータとなり得る。
