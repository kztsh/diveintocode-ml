## [コーディング課題]：1枚の紙を43回折り曲げた時の厚さを計算するプログラム

```python
%matplotlib inline
import time
import matplotlib.pyplot as plt

thickness = 0.00008 # 1枚の紙の厚さ
print("月までの距離：384,400キロメートル")

"""
紙を43回折った時の厚さを計算するコード
"""
# べき乗計算
print("-----べき乗計算-----")
start1 = time.time()
folded_thickness_exp = thickness*(2**43)
print("厚さ： {:.2f}キロメートル".format(folded_thickness_exp/1000))
elapsed_time = time.time() - start1
print("time : {}[s]".format(elapsed_time))

# for文計算
print("-----for文計算-----")
start2 = time.time()
calc_list = []
calc_list.append(thickness) # 折る前の厚さを記録
for i in range(43):
    thickness *= 2
    calc_list.append(thickness) # 折るごとに厚さを記録
print("厚さ： {:.2f}キロメートル".format(thickness/1000))
elapsed_time = time.time() - start2
print("time : {}[s]".format(elapsed_time))
print("厚さ履歴配列の要素数：{}".format(len(calc_list)))

"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
fig, axes = plt.subplots(1, 2, figsize=(15,5))

axes[0].set_title("thickness of folded paper")
axes[0].set_xlabel("number of folds")
axes[0].set_ylabel("thickness[km]")
axes[0].plot(calc_list, label="for sentence calc") 
axes[0].legend(loc="best")

axes[1].set_title("thickness of folded paper (log scale)")
axes[1].set_xlabel("number of folds")
axes[1].set_ylabel("thickness[km] (log_scale)")
axes[1].set_yscale("log")
axes[1].plot(calc_list, label="for sentence calc") 
axes[1].legend(loc="best")
```

## -----考察-----
### 計算時間は大差ないもののべき乗の方が速く、forループよりべき乗の方が計算効率が良さそう。
### 今回、43回も2乗しており、全回数を俯瞰するグラフでは40回弱あたりから急激に増加しているように見える。
### y軸をlog scaleに直すと、べき乗の対数表示なので線形になっていることがわかる。
